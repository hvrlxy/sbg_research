# Heatmap Analysis

This notebook runs code from the heatmap.py file, which is a class crafted to display the student's buy-in of the SBG method through heatmaps.

First, we will import the required header:

```python
from heatmap import *
from heatmap import HeatMap
```

Create a HeatMap instance and display the list of student's IDs.

```python
htm = HeatMap()

print(htm.student_list)
```

# Color code for the heatmap

Apart from the student's activities, the heatmap also reveals information regarding the grade of the student. We can tell whether the student performs well in the course by looking at the color of the heatmap. The reference is listed below:

<ul>
  <li><b>Green</b>: student's score in the A range (A+/-)</li>
  <li><b>Blue</b>: student's score in the B range (B+/-)</li>
  <li><b>Yellow</b>: student's score in the C range (C+/-)</li>
  <li><b>Red</b>: student's score in the D/F range (D+/-/F)</li>
</ul>

The code below displays the heatmap of <b>ALL</b> students in the dataset.

```python
for ID in htm.student_list:
#     print("student ID:",ID)
    htm.plot_student_heatmap(ID)
```

# Filter Heatmap by Grades

There are also an option to plot the heatmap by the grade of the student. We can use this function to compare the "buy-in" of students with different final grades and draw conclusion regarding whether early or constant "buy-in" results in a better outcome.

The code below plot all the heatmap of student with A-range final grade.

```python
#plot heatmaps of all A-range student
htm.plot_heatmap_by_grade('A')
```

The code below plot all the heatmap of student with D/F-range final grade.

```python
#plot heatmaps of all D/F-range student
htm.plot_heatmap_by_grade('D')
htm.plot_heatmap_by_grade('F')
```
